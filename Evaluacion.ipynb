Carga de librerías y archivos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gastos2022 = pd.read_excel("Gastos y costos 20-23.xlsx", sheet_name = "2022")
gastos2023 = pd.read_excel("Gastos y costos 20-23.xlsx", sheet_name = "2023")
```

Tratamiento de nulos 
* Backward Fill
    * Folio
* Clave única (string)
    * MP
    * Poliza
* Mediana
    * TC
* Eliminar
    * Otros

```python
#Copia del Data Frame
gastos2022Limpio = gastos2022.copy()
```

```python
# Backward Fill
    #Folio
gastos2022Limpio["Folio"] = gastos2022["Folio"].fillna(method = "bfill") 
```

```python
#Verificamos
gastos2022Limpio.isnull().sum()
```

```python
#Clave única (string)
    #MP
    #Poliza
gastos2022Limpio["MP"] = gastos2022Limpio["MP"].fillna("NA0")
gastos2022Limpio["Poliza"] = gastos2022Limpio["Poliza"].fillna("NA00-00000")
```

```python
#Verificamos
gastos2022Limpio.isnull().sum()
```

```python
#Mediana
    #TC
gastos2022Limpio["TC"] = gastos2022Limpio["TC"].fillna(gastos2022["TC"].median())
```

```python
#Verificamos
gastos2022Limpio.isnull().sum()
```

```python
#Eliminar
    #Otros
gastos2022Limpio = gastos2022Limpio.drop(["Otros"], axis=1)
```

```python
#Verificamos
gastos2022Limpio.isnull().sum()
```

```python
#Verificamos final
gastos2022Limpio.isnull().sum().sum()
```

Tratamiento de valores atípicos

Se generan dos sub bases: 
* Cualitativas
    * Fecha          
    * Folio          
    * UUID          
    * RFC           
    * Proveedor     
    * TIPO GASTO    
    * Descripción   
    * MP            
    * Tipo          
    * Status    
    * Poliza      
* Cuantitativas
    * TC             
    * Importe        
    * IVA            
    * RET ISR        
    * RET IVA        
    * TOTAL MX       
    * TOTAL SAT      

```python
#Cualitativas
columnas1 = gastos2022Limpio.iloc[:, 0:8]
columnas2 = gastos2022Limpio[["Tipo", "Status", "Poliza"]]
cualitativas2022 = pd.concat([columnas1, columnas2], axis=1)
cualitativas2022
```

```python
#Cuantitativas
cuantitativas2022 = gastos2022Limpio.iloc[:, 8:15]
cuantitativas2022
```

```python
#Diagrama para visualizar valores atípicos
fig = plt.figure(figsize =(15, 8))
gastos2022Limpio.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show()
```

```python
#Tratamiento con desviación estándar
    #Marcamos límites
y = cuantitativas2022
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Observamos valores atípicos del data frame
outliers2022= cuantitativas2022[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers2022
```

```python
#Hacemos el cambio a un nuevo data frame
cuantitativas2022LimpioA= cuantitativas2022[(y<=Limite_Superior)&(y>=Limite_Inferior)]
cuantitativas2022LimpioA
```

```python
#Vemos valores nulos
cuantitativas2022LimpioA.isnull().sum()
```

```python
#Cambiamos los valores nulos que se crearon por la media
cuantitativas2022Final = cuantitativas2022LimpioA.copy()
cuantitativas2022Final = cuantitativas2022Final.fillna(cuantitativas2022LimpioA.mean())
cuantitativas2022Final
```

```python
#Verificamos valores nulos
cuantitativas2022Final.isnull().sum().sum()
```

```python
#Volvemos a observar en diagrama de cajas
fig = plt.figure(figsize =(15, 8))
cuantitativas2022Final.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe Limpio")
plt.show()
```

Creación y visualización de base final

```python
gasto2022Final = pd.concat([cualitativas2022, cuantitativas2022Final], axis=1)
gasto2022Final
```

```python
gasto2022Final.info()
```

```python
#Pasamos a CSV
gasto2022Final.to_csv("Gastos y Costos 2022 Final.xlsx")
```

Tratamiento de nulos

* Clave única (String)
    * MP

* Media 
    *FP
* Eliminar

```python
#Se hace copia del data frame
gastos2023Limpio = gastos2023.copy()
```

```python
#Backward Fill
    #Folio
gastos2023Limpio["Folio"] = gastos2023Limpio["Folio"].fillna(method = "bfill")
```

```python
#Verificamos
gastos2023Limpio.isnull().sum()
```

```python
#Clave única (string)
    #MP
    #Poliza
    #TIPO GASTO
gastos2023Limpio["MP"] = gastos2023Limpio["MP"].fillna("NA0")
gastos2023Limpio["Poliza"] = gastos2023Limpio["Poliza"].fillna("NA00-00000")
gastos2023Limpio["TIPO GASTO"] = gastos2023Limpio["TIPO GASTO"].fillna("DESCONOCIDO")
```

```python
#Verificamos
gastos2023Limpio.isnull().sum()
```

```python
#Media
    #FP
gastos2023Limpio["FP"] = gastos2023Limpio["FP"].fillna(gastos2023Limpio["FP"].mean())
```

```python
#Verificamos
gastos2023Limpio.isnull().sum()
```

```python
#Eliminamos
    #Otros
gastos2023Limpio = gastos2023Limpio.drop(["Otros"], axis=1)
```

```python
#Verificamos
gastos2023Limpio.isnull().sum()
```

```python
#Verificamos completo
gastos2023Limpio.isnull().sum().sum()
```

Tratamiento de valores atípicos

Se generan dos sub bases:
* Cualitativas
    * Fecha          
    * Folio          
    * UUID          
    * RFC           
    * Proveedor     
    * TIPO GASTO    
    * Descripción   
    * MP            
    * Tipo          
    * Status    
    * Poliza      
* Cuantitativas
    *FP
    * TC             
    * Importe        
    * IVA            
    * RET ISR        
    * RET IVA        
    * TOTAL MX       
    * TOTAL SAT   

```python
#Cualitativas
columnas3 = gastos2023Limpio.iloc[:, 0:8]
columnas4 = gastos2023Limpio[["Tipo", "Status", "Poliza"]]
cualitativas2023 = pd.concat([columnas3, columnas4], axis=1)
cualitativas2023
```

```python
#cuantitativas
cuantitativas2023 = gastos2023Limpio.iloc[:, 8:16]
cuantitativas2023
```

```python
#Diagrama de visualización de outliers
fig = plt.figure(figsize =(15, 8))
gastos2023Limpio.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show()
```

```python
#Tratamiento con cuantiles
y=cuantitativas2023

#Encontramos los cuantiles
percentile25=y.quantile(0.25)
percentile75=y.quantile(0.75)
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Visualizamos los valores atípicos
outliers2023= cuantitativas2023[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers2023
```

```python
#Cambiamos valores atípicos por cuantiles
cuantitativas2023LimpioA= cuantitativas2023[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
cuantitativas2023LimpioA
```

```python
#Observamos si hay valores nulos que se generaron
cuantitativas2023LimpioA.isnull().sum()
```

```python
#Cambiamos los nuevos NAs por la media
cuantitativas2023Final=cuantitativas2023LimpioA.copy()
cuantitativas2023Final=cuantitativas2023Final.fillna(cuantitativas2023LimpioA.mean())
cuantitativas2023Final
```

```python
#Verificamos valores nulos
cuantitativas2023Final.isnull().sum().sum()
```

```python
#Se observan en el diagrama de caja nuevamente los outliers
fig = plt.figure(figsize =(15, 8))
cuantitativas2023Final.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe Limpio")
plt.show()
```

Creación y visualización del data frame final

```python
gastos2023Final = pd.concat([cualitativas2023, cuantitativas2023Final], axis=1)
gastos2023Final
```

```python
gastos2023Final.info()
```
